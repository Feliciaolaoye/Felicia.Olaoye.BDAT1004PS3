# Question 1: Occupation

```python
# Step 1: Import the necessary libraries
import pandas as pd











```

```python
# Step 2: Import the dataset from the address
# Assuming the dataset is in a CSV format
file_path = "C:/user.txt"  # Replace with the correct path
users = pd.read_csv(file_path, delimiter='|')  # Adjust the delimiter if needed
```

```python
# Step 4: Discover what is the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
```

```python
# Step 5: Discover the Male ratio per occupation and sort it from the most to the least
male_ratio_per_occupation = users.groupby(['occupation', 'gender'])['gender'].count()
male_ratio_per_occupation = male_ratio_per_occupation.unstack().fillna(0)
male_ratio_per_occupation['Male Ratio'] = male_ratio_per_occupation['M'] / (male_ratio_per_occupation['M'] + male_ratio_per_occupation['F'])
male_ratio_per_occupation = male_ratio_per_occupation['Male Ratio'].sort_values(ascending=False)
```

```python
# Step 6: For each occupation, minimum and maximum ages was calculated
min_max_ages_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
```

```python

# Step 7: For each combination of occupation and sex,the mean age was calculated
mean_age_per_combination = users.groupby(['occupation', 'gender'])['age'].mean()
```

```python

# Step 8: For each occupation,the percentage of women and men was presented
gender_percentage_per_occupation = users.groupby(['occupation', 'gender'])['gender'].count()
gender_percentage_per_occupation = gender_percentage_per_occupation.unstack().fillna(0)
gender_percentage_per_occupation['Total'] = gender_percentage_per_occupation['M'] + gender_percentage_per_occupation['F']
gender_percentage_per_occupation['Female Percentage'] = (gender_percentage_per_occupation['F'] / gender_percentage_per_occupation['Total']) * 100
gender_percentage_per_occupation['Male Percentage'] = (gender_percentage_per_occupation['M'] / gender_percentage_per_occupation['Total']) * 100
gender_percentage_per_occupation = gender_percentage_per_occupation[['Female Percentage', 'Male Percentage']]
```

```python

# Displaying the results
print("Mean Age per Occupation:\n", mean_age_per_occupation)
print("\nMale Ratio per Occupation (sorted):\n", male_ratio_per_occupation)
print("\nMinimum and Maximum Ages per Occupation:\n", min_max_ages_per_occupation)
print("\nMean Age per Combination of Occupation and Sex:\n", mean_age_per_combination)
print("\nPercentage of Women and Men per Occupation:\n", gender_percentage_per_occupation)
```

```python
#step 3
data["Date"] = pd.to_datetime(data[["Yr","Mo","Dy"]].astype(str).agg('-'.join, axis=1))
data = data.drop(columns=["Yr","Mo","Dy"])
data.head()
```

```python
#step 4
data["Date"] = np.where(pd.DatetimeIndex(data["Date"]).year < 2000,data.Date,data.Date - pd.offsets.DateOffset(years=100))
data.head()
```

```python
#step 5
data = data.set_index("Date")
data.index.astype("datetime64[ns]")
```

```python
#step 6
data.isnull().values.ravel().sum()
```

```python
#step 7
data.count().sum()
```

```python
#step 8
pd.DataFrame(data.mean())
```

```python
#step 9
def stats(stats):
    stats = pd.Series(stats)
    Min = stats.min()
    Max = stats.max()
    Mean = stats.mean()
    Std = stats.std()
    res = [Min,Max,Mean,Std]
    indx = ["Min","Max","Mean","Std"]
    res = pd.Series(res,index=indx)
    return res
loc_stats = data.apply(stats)
pd.DataFrame(loc_stats)
```

```python
#step 10
day_stats = data.apply(stats,axis=1)
pd.DataFrame(day_stats.head())
```
